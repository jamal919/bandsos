```python
import pandas as pd
```

```python
df = pd.read_csv('station.in', delim_whitespace=True, header=None, names=['ID', 'Lon', 'Lat', 'Depth', 'Name'])
df.head(30)
```

```python
df['Organization'] = df.Name.apply(lambda x: x.split('_')[1] if len(x.split('_'))==2 else '')
df.Organization.unique()
```

```python
df['Name'] = df.Name.apply(lambda x: x.split('_')[0])
```

```python
df.where(
    df.Organization=="BWDB"
).dropna().drop(
    columns=['Depth']
).reset_index(
    drop=True
).to_csv('bwdb.csv', index=False)
```

```python
df.where(
    df.Organization=="IOC"
).dropna().drop(
    columns=['Depth', 'Organization']
).reset_index(
    drop=True
).to_csv('ioc.csv', index=False)
```

```python
df.where(
    df.Organization=="SOB"
).dropna().drop(
    columns=['Depth', 'Organization']
).reset_index(
    drop=True
).to_csv('sob.csv', index=False)
```

```python
df.where(
    df.Organization=="MPA"
).dropna().drop(
    columns=['Depth', 'Organization']
).reset_index(
    drop=True
).to_csv('mpa.csv', index=False)
```

```python
df.where(
    df.Organization=="CPA"
).dropna().drop(
    columns=['Depth', 'Organization']
).reset_index(
    drop=True
).to_csv('cpa.csv', index=False)
```
